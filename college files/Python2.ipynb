Positional arguments

```
def example(name,place):
  print(name.title()+" is a name")
  print(place.title()+" is a place")
example('ram','trivandrm')
example('trivandrm','ram')
```

keyword arguments

```
def example(name,place):
  print(name.title()+" is a name")
  print(place.title()+" is a place")
example(name='ram',place='trivandrum')
example(place='trivandrum',name='ram')
```

default values

```
def example(name,place='trivandrum'):
  print(name.title()+" is a name")
  print(place.title()+" is a place")
example(name='ram')
```

return values

```
def large_num(a,b):
  if (a>b):
      return a
  else:
      return b
print("Larger number is :",large_num(2,3))
```

Returning a dictionary

```
def build_person(first_name,last_name):
  """Return a dictionary of information about a person"""
  person ={'first':first_name,'last':last_name}
  return person
name=build_person('APJ Abdul','Kalam')
print (name)
```

Passing a list

```
def greet_users(names):
  for name in names:
       msg="Hello,"+ name.title()+"!"
       print (msg)
usernames=['raj','ram','john']
greet_users(usernames)
```

Passing an arbitrary number of arguments

```
def fruits(*fruit):
   print(fruit)
fruits('apple')
fruits('grapes','oranges','mango')
```

Using Arbitrary keyword Arguments

```
def sample(first,last,**user_info):
    profile={}
    profile['first_name']=first
    profile['last_name']=last
    for key,value in user_info.items():
       profile[key]=value
    return profile
user_profile=sample('albert','einstein',location='princeton',field='physics')
print(user_profile)
```

```
#without lambda 
def func(x):return x**3
print(func(5))
```

```
# with lambda
lamb=lambda x:x**3
print(lamb(5))
```

map()

```
#without map
items=[1,2,3,4,5]
squared=[]
for i in items:
  squared.append(i**2)
print(items,squared)
```

```
#with map
items=[1,2,3,4,5]
squared=list(map(lambda x:x**2,items))
print(items)
print(squared)
```

```
def multiply(x):
  return(x*x)
def add(x):
  return(x+x)
funcs=[multiply,add]
for i in range(5):
   value=list(map(lambda x:x(i),funcs))
   print(value)
```

filter()

```
number_list=range(-5,5)
less_than_zero=list(filter(lambda x:x<0,number_list))
print(less_than_zero)
```
